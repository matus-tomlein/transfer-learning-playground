```python
import pandas as pd
```

```python
index = pd.to_datetime([1349720105.21, 1349720105.63, 1349720106.532, 1349720106.92312, 1349720107.642], unit='s')
series = pd.Series([1, 2, 3, 4, 5], index=index)
series
```

```python
resampled = series.resample('100L').ffill().bfill()
resampled
```

```python
resampled['2012-10-08 18:15:05':'2012-10-08 18:15:06']
```

```python
resampled.to_csv()
```

```python
data_frame = pd.DataFrame(data={'A': [1, 2, 3, 4, 5], 'B': [5, 4, 3, 2, 1]}, index=index)

data_frame = data_frame.resample('100L').ffill().bfill()
```

```python
data_frame.to_csv('data.csv')
```

```python
data_frame = pd.DataFrame.from_csv('data.csv')
```
