```python
def plot_accelerometer(device, window_id):
    df = pd.read_pickle(dataset_path + device + '.p')
    df_labels = pd.read_pickle(dataset_path + device + '_labels.p')
    id = df_labels.loc[df_labels.window_id == window_id].index[0]
    df = df.loc[df.id == id].filter(regex='accel_x|accel_y')
    df.plot()
    plt.title(device)
```

```python
plot_accelerometer('xdk_1', chopping_food_window_id)
plot_accelerometer('xdk_2', chopping_food_window_id)
plot_accelerometer('TI SensorTag 604', chopping_food_window_id)
plot_accelerometer('TI SensorTag 690', chopping_food_window_id)
plot_accelerometer('128.237.246.127', chopping_food_window_id)
plot_accelerometer('128.237.247.134', chopping_food_window_id)
plot_accelerometer('Matrix b827ebe6e0f8', chopping_food_window_id)
plot_accelerometer('Matrix b827eb96f31a', chopping_food_window_id)
plot_accelerometer('DialogIoT 591844595', chopping_food_window_id)
plot_accelerometer('DialogIoT 591844599', chopping_food_window_id)
```
