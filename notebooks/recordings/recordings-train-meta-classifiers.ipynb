```python
import pandas as pd
import numpy as np
from tsfresh.transformers import RelevantFeatureAugmenter, FeatureAugmenter
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import Imputer
from sklearn.decomposition import PCA
```

```python
df = pd.DataFrame.from_csv('../../datasets/ESC-50/5_samples.csv')
df_labels = pd.DataFrame.from_csv('../../datasets/ESC-50/5_sample_labels.csv')

y = df_labels['label']
X = pd.DataFrame(index=y.index)

ppl = Pipeline([
    ('fresh', RelevantFeatureAugmenter(column_id='id')),
    ('impute', Imputer()),
    ('scale', StandardScaler()),
    ('clf', RandomForestClassifier())
])

ppl.set_params(fresh__timeseries_container=df)
ppl.fit(X, y)
```

```python
def read_data_frame(name):
    dataset_path = '../../datasets/recordings/'
    return pd.DataFrame.from_csv(dataset_path + name + '.csv')

df_train = read_data_frame('NSH2_iPhone')
df_labels_train = read_data_frame('NSH2_iPhone_labels')

y_train = df_labels_train['label']
X_train = pd.DataFrame(index=y_train.index)
```

```python
ppl.set_params(fresh__timeseries_container=df_train)

y_train_pred = ppl.predict(X_train)
```

```python
df_pred = pd.DataFrame(y_train_pred)
df_pred.index = y_train.index
df_pred.to_csv('../recordings/NSH2_iPhone_ESC_5.csv')
```
